## **01 인공지능 제품, 서비스 조사**

1. 제품(서비스) 개요 
  * AI 스피커 기반 돌봄이웃 스마트 안심케어서비스 구축
  * 양질의 AI 서비스를 위해서는 음성인식을 비롯한 바이오정보의 원본을 기반으로 알고리즘을 개발
  * 사투리 알아 듣는 AI스피커

  ![이미지 설명](https://www.nugu.co.kr/assets/img/device/nugu-grow-device_lg.png)
  ![이미지 설명](https://gigagenie.kt.com/images/main/genieOtherProduct_img01.png)
2. 제품(서비스) 목적
  1. 현대 사회는 초고령사회의 형태를 보이고 있으며, 돌봄이 필요한 취약계층에 대한 돌봄 인프라, 서비스양의 한계발견
  2. 코로나19의 장기화에 따른 비대면 서비스에 대한 필요와 인적, 시간적, 공간적 제약을 넘어서는 혁신적인 서비스가 필요
  3. 독거노인, 장애인, 기저질환자 등 사회적 취약계층이 위급상황에서도 24시간 보호 받을 수 있는 인공지능 돌봄시스템 구축이 요구
3. 제품(서비스)에 사용된 인공지능 기술
  1. 일상생활의 질 향상을 위한 음악, 대화, 날씨정보제공, 뉴스알림, 라디오 청취 서비스, 복약지도, 공공기관 정보 알림 서비스, 안부확인 등
  2. 정서적 지원과 잔존능력 유지를 위한 치매예방 및 우울증 특화프로그램 서비스, 응급상황 시 대응(응급알림, 상담사 연결, 경비업체 및 119를 통한 관제 출동서비스)등
  3. AI 스피커 기반 돌봄이웃 스마트 안심 케어서비스 대상자의 온라인 관제를 위한 관제시스템
  4. 기저질환자, 지체장애인을 대상으로 AI스피커를 활용한 문열림, 조명관리센서 관리 통합연계서비스
